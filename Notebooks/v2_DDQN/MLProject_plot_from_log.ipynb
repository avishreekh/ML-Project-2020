```python
logfile = open('./train_log.txt', 'r')

logs = logfile.readlines()
```

```python
episodes_arr = range(645)
scores_arr = []
average_arr = []
av100_arr = []
for line in logs:
  if line[:8] == 'Episode:':
    linearr = line.split(' ')
    scores_arr.append(float(linearr[5]))
    av100_arr.append(float(linearr[12]))
    average_arr.append(sum(scores_arr)/len(scores_arr))
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.plot(episodes_arr, average_arr, 'r', label='Average score over all episodes')
plt.plot(episodes_arr, scores_arr, 'b', label='Episode score')
plt.plot(episodes_arr, av100_arr, 'g', label='Average score over last 100 episodes')
plt.ylabel('Score', fontsize=18)
plt.xlabel('Steps', fontsize=18)
plt.legend(loc='lower right')
plt.savefig('best_train_plot.png')
```
